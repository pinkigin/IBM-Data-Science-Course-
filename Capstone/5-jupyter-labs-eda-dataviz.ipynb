### TASK  3: Visualize the relationship between success rate of each orbit type


Next, we want to visually check if there are any relationship between success rate and orbit type.


Let's create a `bar chart` for the sucess rate of each orbit


```python
# HINT use groupby method on Orbit column and get the mean of Class column
df=pd.read_csv("https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DS0321EN-SkillsNetwork/datasets/dataset_part_2.csv")

# Group the DataFrame by the 'Category' column

agg_functions = {'Class': ['mean']}
grouped_df = df.groupby('Orbit').agg(agg_functions)


# Flatten the MultiIndex columns and Round values
grouped_df.columns = grouped_df.columns.droplevel()
grouped_df = grouped_df.round(2)
# Rename the mean column to a custom name
grouped_df = grouped_df.rename(columns={'mean': 'Class_mean'})

# Print the resulting DataFrame
print(grouped_df)
```

```python
grouped_df
```

```python


# Create the bar plot
plt.figure(figsize=(10, 6))
ax = sns.barplot(y="Class_mean", x="Orbit", hue='Orbit', data=grouped_df, dodge=False, palette="hsv", legend=False)

# Add labels
ax.set_xlabel("Orbit", fontsize=14)
ax.set_ylabel("Outcome (Success Rate)", fontsize=14)
ax.set_title('Success Rate per Orbit', fontsize=18, pad=20)

# Add labels to the bars
for index, row in grouped_df.iterrows():
    ax.text(index, row.Class_mean, f'{row.Class_mean:.2f}', color='grey', ha="center", va="bottom")

# Show the plot
plt.show()
```

Analyze the ploted bar chart try to find which orbits have high sucess rate.


### TASK  4: Visualize the relationship between FlightNumber and Orbit type


For each orbit, we want to see if there is any relationship between FlightNumber and Orbit type.


```python
# Plot a scatter point chart with x axis to be FlightNumber and y axis to be the Orbit, and hue to be the class value
df=pd.read_csv("https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DS0321EN-SkillsNetwork/datasets/dataset_part_2.csv")
# Rename the class labels
df['Class'] = df['Class'].map({0: 'Failure', 1: 'Successful'})

g = sns.catplot(y="Orbit", x="FlightNumber", hue="Class", data=df, aspect = 2)

# Add labels
g.set_axis_labels("Flight Number", "Orbit", fontsize=14)

# Add title
g.fig.suptitle('Flight Number vs Orbit', fontsize=18)
g.fig.subplots_adjust(top=0.85)  # Adjust the title position

# Show the plot
plt.show()
```

You should see that in the LEO orbit the Success appears related to the number of flights; on the other hand, there seems to be no relationship between flight number when in GTO orbit.


### TASK  5: Visualize the relationship between Payload and Orbit type


Similarly, we can plot the Payload vs. Orbit scatter point charts to reveal the relationship between Payload and Orbit type


```python

# Load the dataset
df = pd.read_csv("https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DS0321EN-SkillsNetwork/datasets/dataset_part_2.csv")

# Rename the class labels
df['Class'] = df['Class'].map({0: 'Failure', 1: 'Successful'})

# Create the figure
plt.figure(figsize=(14, 5))

# Plot the scatter plot
g = sns.scatterplot(x="PayloadMass", y="Orbit", hue="Class", data=df, s=100)

# Add labels
plt.xlabel("Payload Mass (Kg)", fontsize=14)
plt.ylabel("Orbit", fontsize=14)

# Add title
plt.title('Payload vs Orbit', fontsize=16)

# Show the plot
plt.show()
```

### TASK  6: Visualize the launch success yearly trend


You can plot a line chart with x axis to be <code>Year</code> and y axis to be average success rate, to get the average launch success trend. 


The function will help you get the year from the date:


```python

# Plot a scatter point chart with x axis to be FlightNumber and y axis to be the Orbit, and hue to be the class value
df=pd.read_csv("https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DS0321EN-SkillsNetwork/datasets/dataset_part_2.csv")# A function to Extract years from the date 
year=[]
def Extract_year(date):
    for i in df["Date"]:
        year.append(i.split("-")[0])
    return year

df['year']=Extract_year(df["Date"])
df_year_success=df.groupby("year",as_index=False)["Class"].mean()
```

```python

# Plot a line chart with x axis to be the extracted year and y axis to be the success rate


plt.figure(figsize=(14,5))
sns.lineplot(data=df_year_success, x="year", y="Class" )
sns.scatterplot(x="year", y="Class" ,hue="Class",data = df_year_success,s=100 )
# Access the legend and set its title to an empty string
plt.legend(title='Success Rate')

plt.xlabel("Year",fontsize=20)
plt.title('Launch Success Yearly Trend')
plt.ylabel("Success Rate",fontsize=20)
plt.show()
```

You can observe that the success rate since 2013 kept increasing till 2017 (stable in 2014) and after 2015 it started increasing.


## Features Engineering 


By now, you should obtain some preliminary insights about how each important variable would affect the success rate, we will select the features that will be used in success prediction in the future module.


```python
features = df[['FlightNumber', 'PayloadMass', 'Orbit', 'LaunchSite', 'Flights', 'GridFins', 'Reused', 'Legs', 'LandingPad', 'Block', 'ReusedCount', 'Serial']]
features.head()
```

 ### TASK  7: Create dummy variables to categorical columns


Use the function <code>get_dummies</code> and <code>features</code> dataframe to apply OneHotEncoder to the column <code>Orbits</code>, <code>LaunchSite</code>, <code>LandingPad</code>, and <code>Serial</code>. Assign the value to the variable <code>features_one_hot</code>, display the results using the method head. Your result dataframe must include all features including the encoded ones.


```python

# HINT: Use get_dummies() function on the categorical columns
features_one_hot = features

```

```python
features_one_hot = pd.concat([features_one_hot,pd.get_dummies(df['Orbit'])], axis=1)
features_one_hot.drop(['Orbit'], axis = 1,inplace=True)

features_one_hot.head()
```

```python
features_one_hot = pd.concat([features_one_hot,pd.get_dummies(df['LaunchSite'])], axis=1)
features_one_hot.drop(['LaunchSite'], axis = 1,inplace=True)

features_one_hot = pd.concat([features_one_hot,pd.get_dummies(df['LandingPad'])], axis=1)
features_one_hot.drop(['LandingPad'], axis = 1,inplace=True)

features_one_hot = pd.concat([features_one_hot,pd.get_dummies(df['Serial'])], axis=1)
features_one_hot.drop(['Serial'], axis = 1,inplace=True)

features_one_hot.head()
```
